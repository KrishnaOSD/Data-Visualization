```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv("/content/WorldCupMatches.csv")
data
```

```python
print(data)
print(data.columns)
print(data.columns.values)
print(data[20:30])
print(data.head(6))
print(data.tail(20))
print(data["City"])
print(data.describe())
print(data["Year"].sum())
print(data.sort_values(by="Half-time Home Goals"))
```

```python
print(data["Year"].mean())
print(data["Year"].median())
print(data["Year"].mode()[0])
print(data["Year"].max())
print(data["Year"].min())
```

```python
print(data["City"].isnull())#to determine whether an expression contains a Null value
print(data["City"].isnull().sum())#to replace existing NULL values with a specific value.
data["City"].fillna("abcd",inplace=True)
print(data)
```

```python
plt.plot(data["Year"],color="blue",label='id')
plt.legend()#to place a legend on the axis
plt.show()
```

```python
data1=data["Home Team Goals"]
data2=data["Away Team Goals"]
plt.plot(data1,color="red",label="Home Team Goals")
plt.plot(data2,color="blue",label="Away Team Goals")
plt.legend()
plt.show()

```

```python
sns.jointplot(x="Home Team Goals",y="Away Team Goals",data=data)# Identifying Relationships
```

```python
sns.boxplot(data=data,x="Home Team Goals")#to show distributions of numeric data values, especially when you want to compare them between multiple groups.
```

```python
sns.pairplot(data)#enables the visualization of the relationship between each pair of variables in a dataset.
```

```python
 #scatterplot is used to find ratio between two variabels
 sns.scatterplot(x='Home Team Goals',y='Away Team Goals',data=data)
```

```python
sns.histplot(data=data,x='Away Team Goals')# to show distributions of datasets a histogram
```

```python
sns.displot(x='Away Team Goals',data=data)#how to plot distribution
```

```python
#correalation: how more than two categorical or numerical attribute are relatedto each other
corr=data[['Half-time Home Goals','Home Team Goals','Half-time Away Goals']].corr()
sns.heatmap(corr)
```
